#Syntax Analysis of code:

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/zeelpatel7/Measuring-Chatbot-Efficiency/main/Data%20Files/Syntax%20Analysis.csv'
df = pd.read_csv(url)
print(df)
```

```python
df = df.drop('Language', axis=1)
df = df.drop('Code', axis=1)
```

```python
df
```

###Syntax Analysis Heatmap




```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns

data = df

fig, hm = plt.subplots(figsize=(10,10))

hm = sns.heatmap(data=data, square='True', linewidth='0.5')

hm.set_title('Syntax Analysis', weight='bold')
hm.set_xlabel('Chatbots')
hm.set_ylabel('Code')

plt.show()
```
