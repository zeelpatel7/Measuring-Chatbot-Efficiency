```python
import pandas as pd

url = 'https://raw.githubusercontent.com/zeelpatel7/Measuring-Chatbot-Efficiency/main/Data%20Files/Lines%20Of%20Code.csv'
lines = pd.read_csv(url)
print(lines)
```

###CHATGPT Graphical Representation

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler

print(lines['ChatGPT'])
y1= lines['ChatGPT']
x1 = [*range(1,16,1)]

plt.bar(x1, y1, color = "#6A041D")
plt.xlabel("Serial Number")
plt.ylabel("No. of Lines of Code")
plt.title("ChatGPT - Lines of Code")
plt.ylim([0,100])
plt.show()
```
