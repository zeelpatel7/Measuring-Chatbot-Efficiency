#Analysing Speed of Retrieval of code:

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/zeelpatel7/Measuring-Chatbot-Efficiency/main/Data%20Files/Speed%20of%20Code%20Retrieval.csv'
speed = pd.read_csv(url)
print(speed)
```

###CHATGPT Graphical Representation

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler

print(speed['Min Max Scaling'])
X1= speed['Min Max Scaling']
y1 = [*range(1,16,1)]
x1 = X1
scaler1 = MinMaxScaler()

plt.plot(y1, x1, c ="red")
plt.xlabel ("Serial Number")
plt.ylabel("Min Max Scaling of ChatGPT")
plt.ylim([0,1])
plt.xlim([0,15])

plt.show()
```
